## Start scraping

```python
driver = webdriver.Firefox(executable_path=r'C:\Program Files (x86)\geckodriver.exe')
```

```python
url = r"https://www.iea.org/data-and-statistics/data-tools/critical-minerals-policy-tracker" # define url object
```

```python
driver.get(url)
```

```python
# Wait for the page to load completely
time.sleep(5)  # Adjust this sleep time if necessary
```

```python
try:
    # Locate the "All Policies" button and click it
    # Replace the locating strategy if this doesn't work, you might use find_element_by_class_name or another method.
    all_policies_button = driver.find_element_by_partial_link_text("All policies")
    all_policies_button.click()
    
    # Wait for the content to load after clicking
    time.sleep(5)  # Adjust this sleep time if necessary
    
    # Get the page source
    page_source = driver.page_source
finally:
    # Quit the Selenium web driver
    driver.quit()
```

```python
# Parse the page source with Beautiful Soup
soup = bs.BeautifulSoup(page_source, 'html.parser')
```

```python
# Creating list with all tables
tables = soup.find_all('table')
print(tables)
```

```python
# Verifying tables and their classes
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
# Putting each cell into new table going row by row through each column
data = []
for row in table.find_all('tr'):
    row_data = []
    for cell in row.find_all('td'):
        row_data.append(cell.text)
    data.append(row_data)
```

```python
cmp = pd.DataFrame(data)
```

```python
# give names to columns/variables
cmp.columns = ["Region", "Policy", "Start Year", "Policy Type", "Status"]
```
